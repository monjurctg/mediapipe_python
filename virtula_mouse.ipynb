```python
import cv2
import mediapipe as mp
import pyautogui

# Initialize webcam
cap = cv2.VideoCapture(0)
hand_detector = mp.solutions.hands.Hands(max_num_hands=1)
drawing_utils = mp.solutions.drawing_utils
screen_width, screen_height = pyautogui.size()
prev_index_y = None
prev_middle_y = None


while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame = cv2.flip(frame, 1)
    frame_height, frame_width, _ = frame.shape
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    output = hand_detector.process(rgb_frame)
    hands = output.multi_hand_landmarks

    if hands:
        for hand in hands:
            drawing_utils.draw_landmarks(frame, hand, mp.solutions.hands.HAND_CONNECTIONS)
            landmarks = hand.landmark

            # Get index finger tip (id 8) and thumb tip (id 4)
            index_finger = landmarks[8]
            thumb_finger = landmarks[4]
            index_tip = landmarks[8]
            middle_tip = landmarks[12]
            index_y_curr = int(index_tip.y * frame_height)
            middle_y_curr = int(middle_tip.y * frame_height)

            # Convert to screen coordinates
            index_x = int(index_finger.x * frame_width)
            index_y = int(index_finger.y * frame_height)
            screen_x = int(index_finger.x * screen_width)
            screen_y = int(index_finger.y * screen_height)
            # Draw for visual reference
            cv2.circle(frame, (int(index_tip.x * frame_width), index_y_curr), 10, (0, 255, 255), -1)
            cv2.circle(frame, (int(middle_tip.x * frame_width), middle_y_curr), 10, (0, 255, 0), -1)
            # Check if we have previous values
            if prev_index_y is not None and prev_middle_y is not None:
                # Check if both moved upward (y decreased)
                if (prev_index_y - index_y_curr > 40) and (prev_middle_y - middle_y_curr > 40):
                    pyautogui.scroll(150)  # scroll up
                    cv2.putText(frame, "Scrolling Up", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 255), 2)

            # Update previous positions
            prev_index_y = index_y_curr
            prev_middle_y = middle_y_curr

            # Move mouse pointer
            pyautogui.moveTo(screen_x, screen_y)

            # Draw index finger circle
            cv2.circle(frame, (index_x, index_y), 10, (0, 255, 255), -1)

            # Check distance between index and thumb
            thumb_x = int(thumb_finger.x * frame_width)
            thumb_y = int(thumb_finger.y * frame_height)
            cv2.circle(frame, (thumb_x, thumb_y), 10, (255, 0, 0), -1)

            distance = ((thumb_x - index_x)**2 + (thumb_y - index_y)**2) ** 0.5
            if distance < 20:
                pyautogui.click()
                cv2.putText(frame, "Click", (index_x, index_y - 20), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    # Show camera feed
    cv2.imshow("Virtual Mouse", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
